# DisposableConnection

A disposable connection can be created for any disposable object located in frontend once it has been found. The disposable connection makes the frontend object the `base` for the generic methods. 

Some classes already extend `DisposableConnection` to to add more features, such as commands, command pallet and launchers. These will automatically replace the base class when the `ID_PREFIX` of the subclass 

|  Prefix                    |  Class                             |
| ---------------------------| ---------------------------------- |
|ipylab command 	         | CommandConnection                  |
|ipylab command pallet       | 			 CommandPalletConnection  |
|ipylab launcher 			 | LauncherConnection                 |

## Making a connection
